```python
##Do some Outlier study and remove the outliers appropriately.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
col=['index','price','number_available_in_stock','number_of_reviews','number_of_answered_questions','average_review_rating']
for c in col:
    per25=df[c].quantile(0.25)
    per75=df[c].quantile(0.75)
    iqr=per75-per25
    upperlimit=per75+(1.5*iqr)
    lowerlimit=per25-(1.5*iqr)
    A=df[(df[c]>lowerlimit) & (df[c]<upperlimit)]
    plt.figure()
    sns.boxplot(y=c,data=A)
```
